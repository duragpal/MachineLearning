# Combining DataSets:

1) concat
2) merge
3) join

```python
data_cust = pd.DataFrame({"customerID":['101','102','103','104'], 
                        'category': ['Medium','Medium','High','Low'],
                        'first_visit': ['yes','no','yes','yes'],
                        'sales': [123,52,214,663]},index=[0,1,2,3])

data_cust_new = pd.DataFrame({"customerID":['101','103','104','105'], 
                    'distance': [12,9,44,21],
                    'sales': [123,214,663,331]},index=[4,5,6,7])
```

```python
data_cust
```

```python
data_quarters
```

```python
data_cust_new
```

```python
pd.concat([data_cust,data_cust_new],axis=0)
```

```python
pd.concat([data_cust,data_cust_new],axis=1)
```

```python
pd.merge(data_cust,data_cust_new,how="outer",on='customerID') #on = column in both datasets
```

```python
pd.merge(data_cust,data_cust_new,how="inner",on='customerID') #gives only information of common entries in 'on' column
```

```python
pd.merge(data_cust,data_cust_new,how="right",on='customerID') #all columns of right dataset
```

```python
data_quarters = pd.DataFrame({'Q1': [101,102,103],
                              'Q2': [201,202,203]},
                               index=['I0','I1','I2'])

data_quarters_new = pd.DataFrame({'Q3': [301,302,303],
                                  'Q4': [401,402,403]},
                               index=['I0','I2','I3'])
```

```python
data_quarters
```

```python
data_quarters_new
```

```python
data_quarters.join(data_quarters_new,how="right") #outer, inner, left, right
```
