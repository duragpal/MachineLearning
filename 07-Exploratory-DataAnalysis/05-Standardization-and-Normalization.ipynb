# Standardization

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std;
    normalized_data.append(z_score)
    
```

```python
normalized_data
```

```python
sns.histplot(total_bill),sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(df[['total_bill','tip']])
```

```python
scalar.transform(df[['total_bill','tip']])
```
